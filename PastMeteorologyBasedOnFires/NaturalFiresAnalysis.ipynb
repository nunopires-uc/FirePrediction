```python
import os
import pandas as pd

current_dir = os.getcwd()
root_dir = os.path.dirname(os.path.dirname(current_dir))
print(root_dir)
_RAWFIRESPATH = f"{root_dir}/FirePrediction/Dataset/AllYearsCombined.csv"
```

```python
df = pd.read_csv(_RAWFIRESPATH)
```
