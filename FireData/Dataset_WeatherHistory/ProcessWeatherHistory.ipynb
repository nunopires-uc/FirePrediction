```python
df = pd.read_csv('CACHE_3MonthsNoFire/batch_0.csv')

data_list = []

horas = ast.literal_eval(df['hourly.time'].iloc[0])

csv_files = glob.glob('CACHE_3MonthsNoFire/*.csv')

csv_files_sorted = sorted(csv_files, key=lambda x: int(x.split('_')[-1].split('.')[0]))

print(csv_files_sorted)

nf = 66

for filename in csv_files_sorted:
    df = pd.read_csv(filename)
    if(filename == 'CACHE_3MonthsNoFire/batch_0.csv'):
        start = 66
    else:
        start = 0
    for f in range(start, len(df)):
        for j in range(len(horas)):
            print(f"{f}:{j}")
            temp_data = {
                    'latitude': df['latitude'].iloc[f],
                    'longitude': df['longitude'].iloc[f],
                    'altitude': df['elevation'].iloc[f],
                }
            for i in range(3, len(df.columns)-1):
                temp_data[df.columns[i]] = ast.literal_eval(df[df.columns[i]].iloc[f])[j]

                
            data_list.append(temp_data)

        new_data = pd.DataFrame(data_list)

        new_data.to_csv(f'OUTPUTNOFIRE/fw_{nf}.csv', index=False)
        nf += 1
        data_list = []
```

```python
import pandas as pd
import glob
import re

def sort_key(file_path):
    match = re.search(r'\d+', file_path)
    if match:
        return int(match.group())
    else:
        return 0

csv_files = glob.glob('OUTPUT/*.csv')
csv_files.sort(key=sort_key)
print(csv_files)


for file in csv_files:
    df = pd.read_csv(file)
    print(file)
    if 'index' not in df.columns:
        new_index = range(len(df)-1, -1, -1)
        df.index = new_index
        df.to_csv(file, index=True, index_label='index')
    else:
        print(f"File {file} already has an index column.")
```
