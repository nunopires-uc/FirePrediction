```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import numpy as np


df = pd.read_csv('AveragesByDay/daily_average.csv')

print(df.columns)

cols = ['day', 'location', 'index', 'Unnamed: 0','hourly.snow_depth', 'index.1']

df = df.drop(cols, axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df.drop('fwix', axis=1)
y = df['fwix']

model = RandomForestRegressor(random_state=145521151)
model.fit(X, y)

importances = pd.DataFrame({'feature':X.columns,'importance':np.round(model.feature_importances_,3)})
importances = importances.sort_values('importance',ascending=False)

print("\nFeature Importances:\n", importances)
```

```python
correlations = df.corr()['fwix'].sort_values(ascending=False)
print("\nCorrelations with fwix:\n", correlations)
```
