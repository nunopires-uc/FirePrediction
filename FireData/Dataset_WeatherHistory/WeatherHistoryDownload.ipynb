```python
with open("urls.txt", "r") as file:
    urls = file.readlines()

print(urls)
print(len(urls))


for u in range(18, len(urls)):
    response = requests.get(urls[u].strip())
    print(response.json())
    print(response.status_code)

    if((response.status_code == 200)):
        data = response.json()
        df = pd.json_normalize(data)
        df.to_csv(f"CACHE_3Months/batch_{str(u)}.csv", index=False)
    else:
        print(u)
        break
```

```python
import glob

csv_files = glob.glob('CACHE_3Months/*.csv')

for filename in csv_files:
    dfires = pd.read_csv(filename)

    column_names = dfires.columns.tolist()
    print(column_names)

    column_names = dfires.columns.tolist()

    columns_to_remove = [col for col in column_names if 'hourly_units' in col]
    columns_to_remove.extend(['generationtime_ms', 'utc_offset_seconds', 'timezone', 'timezone_abbreviation'])

    dfires = dfires.drop(columns=columns_to_remove)

    updated_column_names = dfires.columns.tolist()
    print(updated_column_names)
    print(len(updated_column_names))

    dfires.to_csv(f"{filename}", index=None)
```

```python
df = pd.read_csv('CACHE_3Months/batch_0.csv')

data_list = []

horas = ast.literal_eval(df['hourly.time'].iloc[0])

csv_files = glob.glob('CACHE_3Months/*.csv')

csv_files_sorted = sorted(csv_files, key=lambda x: int(x.split('_')[-1].split('.')[0]))

print(csv_files_sorted)

nf = 0

for filename in csv_files_sorted:
    df = pd.read_csv(filename)
    for f in range(len(df)):
        for j in range(len(horas)):
            print(f"{f}:{j}")
            temp_data = {
                    'latitude': df['latitude'].iloc[f],
                    'longitude': df['longitude'].iloc[f],
                    'altitude': df['elevation'].iloc[f],
                }
            for i in range(3, len(df.columns)-1):
                temp_data[df.columns[i]] = ast.literal_eval(df[df.columns[i]].iloc[f])[j]

                
            data_list.append(temp_data)

        new_data = pd.DataFrame(data_list)

        new_data.to_csv(f'OUTPUT/fw_{nf}.csv', index=False)
        nf += 1
        data_list = []
        horas = ast.literal_eval(df['hourly.time'].iloc[f])
```
