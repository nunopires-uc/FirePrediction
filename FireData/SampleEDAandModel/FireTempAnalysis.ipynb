```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cdf = pd.read_csv('combined.csv')
df2019 = cdf[cdf['region'] == 1]
df2022 = cdf[cdf['region'] == 2]
```

```python
'''hourly.temperature_2m', 'hourly.dew_point_2m', 'hourly.precipitation',
       'hourly.weather_code', 'hourly.pressure_msl', 'hourly.surface_pressure',
       'hourly.cloud_cover', 'hourly.cloud_cover_mid',
       'hourly.cloud_cover_high', 'hourly.wind_speed_10m',
       'hourly.wind_direction_10m', 'hourly.soil_temperature_7_to_28cm',
       'hourly.soil_temperature_28_to_100cm',
       'hourly.soil_temperature_100_to_255cm', 'hourly.soil_moisture_0_to_7cm',
       'hourly.soil_moisture_7_to_28cm', 'hourly.soil_moisture_28_to_100cm',
       'hourly.soil_moisture_100_to_255cm', 'hourly.diffuse_radiation
'''

tail = 18
last_12 = df2022.tail(tail)
last_12x = df2019.tail(tail)

fig, axs = plt.subplots(1, 2, figsize=(20,6))


axs[0].plot(list(last_12['hourly.time']), np.array(last_12['hourly.temperature_2m']).reshape(-1), marker='o', label='hourly.temperature_2m')
axs[0].plot(list(last_12['hourly.time']), np.array(last_12['hourly.relative_humidity_2m']).reshape(-1), marker='o', label='hourly.relative_humidity_2m')
axs[0].plot(list(last_12['hourly.time']), np.array(last_12['hourly.dew_point_2m']).reshape(-1), marker='o', label='hourly.dew_point_2m')
axs[0].set_title('2022')
axs[0].set_xlabel('Hourly Time')
axs[0].set_ylabel('Value')
axs[0].grid(True)
axs[0].legend()
axs[0].set_xticklabels(list(last_12['hourly.time']), rotation=90)


axs[1].plot(list(last_12x['hourly.time']), np.array(last_12x['hourly.temperature_2m']).reshape(-1), marker='o', label='hourly.temperature_2m')
axs[1].plot(list(last_12x['hourly.time']), np.array(last_12x['hourly.relative_humidity_2m']).reshape(-1), marker='o', label='hourly.relative_humidity_2m')
axs[1].plot(list(last_12x['hourly.time']), np.array(last_12x['hourly.dew_point_2m']).reshape(-1), marker='o', label='hourly.dew_point_2m')

axs[1].set_title('2019')
axs[1].set_xlabel('Hourly Time')
axs[1].set_ylabel('Value')
axs[1].grid(True)
axs[1].legend()
axs[1].set_xticklabels(list(last_12x['hourly.time']), rotation=90)

plt.tight_layout()  # Adjust the layout to prevent overlaps
plt.show()
```
