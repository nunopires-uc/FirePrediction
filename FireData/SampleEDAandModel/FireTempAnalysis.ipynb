```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
```

```python
df2015 = pd.read_csv('source2/2015byHourFWIXFFMCFireHour.csv')
df2019 = pd.read_csv('source2/2019byHourFWIXFFMCFireHour.csv')
df2022 = pd.read_csv('source2/2022byHourFWIXFFMCFireHour.csv')
```

```python
'''
hourly.temperature_2m,hourly.relative_humidity_2m,hourly.dew_point_2m,hourly.apparent_temperature,hourly.precipitation,
hourly.rain,hourly.snowfall,hourly.snow_depth,hourly.weather_code,hourly.pressure_msl,
hourly.surface_pressure,hourly.cloud_cover,hourly.cloud_cover_low,
hourly.cloud_cover_mid,hourly.cloud_cover_high,
hourly.et0_fao_evapotranspiration,hourly.vapour_pressure_deficit,

hourly.wind_speed_10m,hourly.wind_speed_100m,hourly.wind_direction_10m,
hourly.wind_direction_100m,hourly.wind_gusts_10m,hourly.soil_temperature_0_to_7cm,

hourly.soil_temperature_7_to_28cm,hourly.soil_temperature_28_to_100cm,
hourly.soil_temperature_100_to_255cm,hourly.soil_moisture_0_to_7cm,


hourly.soil_moisture_7_to_28cm,hourly.soil_moisture_28_to_100cm,
hourly.soil_moisture_100_to_255cm,

hourly.is_day,hourly.sunshine_duration,


hourly.shortwave_radiation,hourly.direct_radiation,
hourly.diffuse_radiation,

hourly.direct_normal_irradiance,hourly.global_tilted_irradiance,


,,,
,hourly.direct_normal_irradiance_instant,
hourly.global_tilted_irradiance_instant,

hourly.terrestrial_radiation
hourly.direct_radiation_instant
hourly.shortwave_radiation_instant
hourly.terrestrial_radiation_instant
hourly.diffuse_radiation_instant
'''

def thin_labels(ax, n):
    for index, label in enumerate(ax.xaxis.get_ticklabels()):
        if index % n != 0:
            label.set_visible(False)
days = 3
tail = (24*days)+1
last_2022 = df2022.tail(tail)
last_2019 = df2019.tail(tail)
last_2015 = df2015.tail(tail)

var1 = 'hourly.terrestrial_radiation'
var2 = 'hourly.direct_radiation_instant'
var3 = 'hourly.shortwave_radiation_instant'
var4 = 'hourly.terrestrial_radiation_instant'
var5 = 'hourly.diffuse_radiation_instant'

name = (var1 + '_' + var2 + '_' + var3).replace('hourly.', '')

fig, axs = plt.subplots(figsize=(20,18))
gs = gridspec.GridSpec(2, 2)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax3 = plt.subplot(gs[1, :])

ax1.plot(list(last_2022['hourly.time']), np.array(last_2022[f'{var1}']).reshape(-1), marker='o', label=f"{var1.replace('hourly.', '')}")
ax1.plot(list(last_2022['hourly.time']), np.array(last_2022[f'{var2}']).reshape(-1), marker='o', label=f"{var2.replace('hourly.', '')}")
ax1.plot(list(last_2022['hourly.time']), np.array(last_2022[f'{var3}']).reshape(-1), marker='o', label=f"{var3.replace('hourly.', '')}")
ax1.plot(list(last_2022['hourly.time']), np.array(last_2022[f'{var4}']).reshape(-1), marker='o', label=f"{var4.replace('hourly.', '')}")
ax1.plot(list(last_2022['hourly.time']), np.array(last_2022[f'{var5}']).reshape(-1), marker='o', label=f"{var5.replace('hourly.', '')}")
ax1.set_title('2022')
ax1.set_xlabel('Hourly Time')
ax1.set_ylabel('Value')
ax1.grid(True)
ax1.legend()
ax1.set_xticklabels(list(last_2022['hourly.time']), rotation=90)

ax2.plot(list(last_2019['hourly.time']), np.array(last_2019[f'{var1}']).reshape(-1), marker='o', label=f"{var1.replace('hourly.', '')}")
ax2.plot(list(last_2019['hourly.time']), np.array(last_2019[f'{var2}']).reshape(-1), marker='o', label=f"{var2.replace('hourly.', '')}")
ax2.plot(list(last_2019['hourly.time']), np.array(last_2019[f'{var3}']).reshape(-1), marker='o', label=f"{var3.replace('hourly.', '')}")
ax2.plot(list(last_2019['hourly.time']), np.array(last_2019[f'{var4}']).reshape(-1), marker='o', label=f"{var4.replace('hourly.', '')}")
ax2.plot(list(last_2019['hourly.time']), np.array(last_2019[f'{var5}']).reshape(-1), marker='o', label=f"{var5.replace('hourly.', '')}")
ax2.set_title('2019')
ax2.set_xlabel('Hourly Time')
ax2.set_ylabel('Value')
ax2.grid(True)
ax2.legend()
ax2.set_xticklabels(list(last_2019['hourly.time']), rotation=90)

ax3.plot(list(last_2015['hourly.time']), np.array(last_2015[f'{var1}']).reshape(-1), marker='o', label=f"{var1.replace('hourly.', '')}")
ax3.plot(list(last_2015['hourly.time']), np.array(last_2015[f'{var2}']).reshape(-1), marker='o', label=f"{var2.replace('hourly.', '')}")
ax3.plot(list(last_2015['hourly.time']), np.array(last_2015[f'{var3}']).reshape(-1), marker='o', label=f"{var3.replace('hourly.', '')}")
ax3.plot(list(last_2015['hourly.time']), np.array(last_2015[f'{var4}']).reshape(-1), marker='o', label=f"{var3.replace('hourly.', '')}")
ax3.plot(list(last_2015['hourly.time']), np.array(last_2015[f'{var5}']).reshape(-1), marker='o', label=f"{var5.replace('hourly.', '')}")
ax3.set_title('2015')
ax3.set_xlabel('Hourly Time')
ax3.set_ylabel('Value')
ax3.grid(True)
ax3.legend()
ax3.set_xticklabels(list(last_2015['hourly.time']), rotation=90)

thin_labels(ax1, 4)
thin_labels(ax2, 4)
thin_labels(ax3, 4)

plt.tight_layout()
plt.savefig(f'{days}_{name}.png')
plt.show()
```
