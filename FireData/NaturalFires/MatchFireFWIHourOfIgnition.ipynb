```python
df = pd.read_csv('NaturalFiresTOPOGRAPHY_mean_elev.csv')
for i in range(len(df)):
    j = i + 1
    df_location = dfhist[dfhist['location'] == j]

    df_location['hourly.time'] = pd.to_datetime(df_location['hourly.time'])

    time = pd.to_datetime(df.loc[i, "hourly.time"])

    df_time = df_location[df_location['hourly.time'] == time]

    df_time = df_time[['fwix', 'ffmc', 'dmc', 'dc', 'isi', 'bui']]

    # Check if df_time is not empty
    if not df_time.empty:
        df.loc[i, ['fwix', 'ffmc', 'dmc', 'dc', 'isi', 'bui']] = df_time.values[0]

df.to_csv('NaturalFiresTOPOGRAPHY_mean_elev_FWI.csv', index=False)
```
