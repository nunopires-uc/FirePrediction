```python
df = pd.read_csv('Samples/CopernicusToOpenMeteo/s2015d.csv')

num_rows = df.shape[0]
print(num_rows)

for r in range(num_rows):
    horas = ast.literal_eval(df['hourly.time'].iloc[r])

    # Initialize an empty list to collect data
    data_list = []

    for j in range(len(horas)):
        temp_data = {
                'latitude': df['latitude'].iloc[r],
                'longitude': df['longitude'].iloc[r],
                'altitude': df['elevation'].iloc[r],
            }
        for i in range(3, len(df.columns)):
            # Create a dictionary for the current line
            temp_data[df.columns[i]] = ast.literal_eval(df[df.columns[i]].iloc[0])[j]
            
        # Append the dictionary to the list
        data_list.append(temp_data)

    # Create a DataFrame from the list of dictionaries
    new_data = pd.DataFrame(data_list)

    # Save the DataFrame to a CSV file
    new_data.to_csv(f'Samples/CopernicusToOpenMeteo/s2015d{r}.csv', index=False)
```
