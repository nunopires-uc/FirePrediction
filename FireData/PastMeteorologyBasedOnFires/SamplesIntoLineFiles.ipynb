```python
import pandas as pd
import ast

# Assuming df and horas are defined elsewhere in your code

df = pd.read_csv('Samples/CopernicusToOpenMeteo/nearestStationInfod.csv')
horas = ast.literal_eval(df['hourly.time'].iloc[0])

# Initialize an empty list to collect data
data_list = []

for j in range(len(horas)):
    temp_data = {
            'latitude': df['latitude'].iloc[0],
            'longitude': df['longitude'].iloc[0],
            'altitude': df['elevation'].iloc[0],
        }
    for i in range(3, len(df.columns)):
        # Create a dictionary for the current line
        temp_data[df.columns[i]] = ast.literal_eval(df[df.columns[i]].iloc[0])[j]
        
    # Append the dictionary to the list
    data_list.append(temp_data)

# Create a DataFrame from the list of dictionaries
new_data = pd.DataFrame(data_list)

# Save the DataFrame to a CSV file
new_data.to_csv('Samples/CopernicusToOpenMeteo/nearestStationInfodd.csv.csv', index=False)
```
