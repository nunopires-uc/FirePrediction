```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
import pandas as pd

# Replace the following values with your actual data
username = "postgres"
password = "postgres"
hostname = "localhost"
database_name = "trees"

engine = create_engine(f'postgresql://{username}:{password}@{hostname}/{database_name}')
Session = sessionmaker(bind=engine)
session = Session()
```

```python
_year = 2023
dfFires = pd.read_csv(f"DatasetWTrees/PreviousVersions/{_year}_checklocality2.csv")
dfTreesDRP = pd.read_csv('TreesPortugueseTerritoryDropped.csv')
```

```python
dfFires.to_sql('dfFires', engine, if_exists='replace', method='multi', index=False)
dfTreesDRP.to_sql('dfTreesDRP', engine, if_exists='replace', method='multi', index=False)
```

```python
from sqlalchemy import text

with engine.connect() as connection:
    connection.execute(text("CREATE INDEX dfFires_gix ON dfFires USING GIST (geom);"))
    connection.execute(text("CREATE INDEX dfTreesDRP_gix ON dfTreesDRP USING GIST (geom);"))
```

```python
from sqlalchemy import inspect

inspector = inspect(engine)

if 'dfFires' in inspector.get_table_names():
    with engine.connect() as connection:
        connection.execute(text("CREATE INDEX dfFires_gix ON dfFires USING GIST (geom);"))
else:
    print("Table dfFires does not exist")

if 'dfTreesDRP' in inspector.get_table_names():
    with engine.connect() as connection:
        connection.execute(text("CREATE INDEX dfTreesDRP_gix ON dfTreesDRP USING GIST (geom);"))
else:
    print("Table dfTreesDRP does not exist")
```
